```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)

plt.title("Sin wave")
plt.xlabel("Angle")
plt.ylabel("Sin Value")
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z,'g--')
plt.show()
```

```python
# bar chart
languages=['English','Hindi','Spanish','chinese']
people=[100,200,300,400]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
plt.xlabel("Languages")
ax.bar(languages,people)
plt.show()
```

```python
# pie chart 
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])

ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
# scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)

fig3=plt.figure()
ax=fig3.add_axes([0,0,1,1])
ax.scatter(x,y,color='green')
ax.scatter(x,z,color='red')
plt.show()
```
