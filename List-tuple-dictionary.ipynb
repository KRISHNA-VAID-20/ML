Types of Objects in Python:
1. Immutable Objects
2. Mutable Objects

Immutable Objects:
1. int
2. float
3. string
4. bool
5. tuple

Mutable Objects:
1. List
2. Set 
3. Dictionary

List

```python
# list should be included in the square brackets
my_list = [1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
# lists can have multiple data types
my_list = [2, 3, 1.8, 'English', True]
print(my_list)
```

Lists are Mutable --> Changeable

```python
# add elements to a list
my_list = [2, 3, 1.8, 'English', True]
my_list.append(6)
print(my_list)
```

```python
# print elements of a list using their index
print(my_list[0])
print(my_list[2])
```

```python
# lists allow duplicate values
list_1 = [1,2,3,4,5,12,2,3]
print(list_1)
```

```python
print(len(list_1))
```

```python
# initiating an empty list
list_2 = []
print(list_2)
```

```python
list_2.append(5)
print(list_2)
```

```python
# delete an item in a list
list_2 = [2, 3, 1.8, 'English', True, 6]
print(list_2)

del list_2[2]
print(list_2)
```

```python
# join two lists
list_3 = [1,2,3,4,5]
list_4 = [6,7,8,9,10]

list_5 = list_3 + list_4
print(list_5)
```

Tuple

```python
tuple_1 = (2,3,4,5)
print(tuple_1)
type(tuple_1)
```

```python
# tuple allows multiple data types
tuple_2 = (1,2,3.5, 'Machine Learning', False)
print(tuple_2)
```

```python
# converting  a list to a tuple

my_list = [3,4,5,6]
print(my_list)

my_tuple = tuple(my_list)
print(my_tuple)
```

```python
print(my_tuple[0])
print(my_tuple[1])
```

Tuples are immutable --> Unchangeable

```python
my_tuple.append(6)
```

```python
print(len(my_tuple))
```

Set

```python
# set --> Curly brackets
my_set = {1,2,3,4,5}
print(my_set)
type(my_set)
```

```python
print(my_set[0])
```

```python
# convert a list to a set
list_5 = [4,5,6,7,8]

x = set(list_5)
print(x)
```

```python
# set does not allow duplicate values
set_3 = {1,2,3,4,5,1,2,3}
print(set_3)
```

Dictionary

Key-Value Pair

```python
my_dictionary = {'name':'David','age':30,'country':'India'}
print(my_dictionary)
type(my_dictionary)
```

```python
print(my_dictionary['name'])
print(my_dictionary['age'])
print(my_dictionary['country'])
```

```python
# dictionary does not allow duplicate values
dictionary_2 = {'name':'David','age':30,'country':'India','name':'David','age':30,'country':'India'}
print(dictionary_2)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load dataset
df = pd.read_csv("housing.csv") # Replace with your CSV file
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)
print("Correlation Matrix:\n", corr_matrix)
# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
target = 'MEDV'
# Get correlations with target
correlations = corr_matrix[target].sort_values(ascending=False)
print("\nCorrelation of features with Housing Price (MEDV):\n", correlations)
# Identify strong correlations (threshold = 0.5)
strong_corr = correlations[abs(correlations) > 0.5]
print("\nStrong correlations with MEDV:\n", strong_corr)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
# Assuming 'CHAS' or 'ZN' is categorical (replace with your datasets categorical variable)
# Example: Compare 'MEDV' across different 'CHAS' (Charles River dummy variable)
model = ols('MEDV ~ C(CHAS)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nANOVA Results:\n", anova_table)
```
